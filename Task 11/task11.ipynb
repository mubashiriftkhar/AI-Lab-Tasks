Data Pre-Processing

```python
df.info()
```

```python
df['Datetime']=pd.to_datetime(df['Datetime'])
df['Datetime']=df['Datetime'].astype('int64')//10**9
df['TotalPowerConsumption']=(df['PowerConsumption_Zone1']
                                        +df['PowerConsumption_Zone2']
                                        +df['PowerConsumption_Zone3'])
df.head()
```

```python
df2=df.copy()
def convertfloatToInt():
    for i in df2:
        df2[i]=df2[i].astype('int64')
convertfloatToInt()
df2.head()        
```

Trian/SplitTest

```python
X=df2[['Datetime','Temperature','Humidity']]
Y=df2['TotalPowerConsumption']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4)
```

Model Training

```python
model=SVR()
model.fit(X_train,Y_train)
```

```python
predictions=model.predict(X_test)
```

```python
score=model.score(X_test,Y_test)
score
```
