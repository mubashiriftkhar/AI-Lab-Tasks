# Train-Test Split

```python
X=df_encoded.drop(['price','statezip','date','country','street'],axis='columns')
Y=df_encoded['price']
X_train, X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.4, shuffle=False)
```

# Model Training using LinearRegression

```python
print("-- Training using Linear Regressions on Training Data --")
print("-- Parameters & Values: ", end='')

# model_svc = SVC(gamma='auto', random_state=0)
# model_svc.fit(train_x, np.ravel(train_y))

model=train()
model.fit(X_train, Y_train)

print(model)
```
